```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [22, 34, 15, 10, 40, 47, 1]
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x, y, color="red", linewidth=2, linestyle="dotted")
```

### Format String in Plot

```python
plt.plot(x, y, 'r+--', alpha=0.7)
```

### Axes Labels, Legends, Grid

```python
days = [1, 2, 3, 4, 5, 6, 7]
min_t = [25, 25, 20, 22, 25, 21, 20]
max_t = [43, 45, 35, 33, 36, 30, 40]
avg_t = [30, 30, 27, 27, 31, 25, 30]
```

```python
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend()
plt.grid()
```

### Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
plt.ylabel("Revenue($Bln)")
plt.xlabel("Companies")
plt.bar(company, revenue)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Stocks")
plt.bar(xpos - 0.2, revenue, width=0.4 ,label="Revenue")
plt.bar(xpos + 0.2, profit, width=0.4, label="Profit")
plt.legend()
```

### Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 179]
blood_sugar_women = [123, 55, 100, 130, 159, 98, 33, 105, 125, 60, 47, 92, 18]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150] , rwidth=0.95, color=['g', 'y'], label=['men', 'women'])
plt.legend()
plt.xlabel("Sugar Range")
plt.ylabel("Frequency")
```

### Pie Chart

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone Bill", "Gas", "Others"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', explode=[0, 0, 0.1, 0.1, 0])
plt.show()
```
